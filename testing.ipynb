```python
bond_graph = {}
for key in rtfs['POPE']['bond'].keys():
    a1, a2 = key.split('-')
    try: 
        bond_graph[a1].add(a2)
    except KeyError:
        bond_graph.update({a1: {a2}})
        
bond_graph
```

```python
edges = []
for node in bond_graph:
    for neighbor in bond_graph[node]:
        if any(x in edges for x in [{neighbor, node}, {node, neighbor}]):
            pass
        else:
            edges.append({node, neighbor})
edges
```
